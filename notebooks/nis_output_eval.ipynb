```python
import pandas as pd
from glob import glob
import os
import matplotlib.pyplot as plt
```

```python
data_dict = { 'radar': {}, 'lidar': {}} # A container for the radar/lidar NIS (for each of the tracked car)

for data_file in glob('../build/run-6-2-13:11:32/*.csv'):
    car, sensor = data_file.split(os.sep)[-1].rstrip('.csv').split('_')
    if sensor in data_dict:
        data_dict[sensor][car] = pd.read_csv(data_file).values
    else:
        data_dict[sensor] = {car: pd.read_csv(data_file).valus}
```

## RADAR NIS (Normalized Innovation Squared)

```python
plt.figure(figsize=(20,  5))
for car, data in data_dict['radar'].items():
    plt.plot(data, label=f' NIS for {car}')
plt.axhline(7.82, color='red', ls='--', label = 'chi square 95%') # X chi
plt.grid()
plt.legend()
```

## LIDAR NIS (Normalized Innovation Squared)

```python
plt.figure(figsize=(20,  5))
for car, data in data_dict['lidar'].items():
    plt.plot(data, label=f' NIS for {car}')
plt.axhline(5.99, color='red', ls='--', label = 'chi square 95%')
plt.grid()
plt.legend()
```
